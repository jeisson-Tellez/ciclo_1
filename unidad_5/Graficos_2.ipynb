Aqui se procesaran datos con contenedores internos y no exportados de portales externos, el contenedor escogido para el ejemplo es un diccionario, este se va transformar en un dataframe

```python
import pandas as pd
df = pd.DataFrame({
    'name':['john','mary','peter','jeff','bill','lisa','jose'],
    'age':[23,78,22,19,45,33,20],
    'gender':['M','F','M','M','M','F','M'],
    'state':['california','dc','california','dc','california','texas','texas'],
    'num_children':[2,0,0,3,2,1,4],
    'num_pets':[5,1,0,5,2,2,3]})

print(df)
```

Grafico de dispersion que analiza cual es el numero de mascotas y el numero de hijos de la persona, se le dio un valor especifico a x,y; se eligio el color de los puntos

```python
df.plot.scatter(x="num_children", y="num_pets", color="black")
```

Esta grafica muestra la informacion en barras, donde a traves del argumento kind=, de la funcion plot, podemos escoger el tipo de  grafica, de igual manera podemos asignarle a x,y valores con etiquetas de las colunmas del dataframe, finalmente darle un color

```python
df.plot(kind="bar",x= "name", y="age", color="green" )
```

Grafica de lineas que evidencia el nombre y el numero de mascotas.

```python
import matplotlib.pyplot as plt
ax=plt.gca()
df.plot.line(x= "name", y="num_children", color="green", ax=ax)
df.plot.line(x= "name", y="num_pets", color="purple", ax=ax)
```

la siguiente grafica, df.groupby("state") agrupa por estados los datos-> cre un dataframe con los datos agrupados,combinados con ["name"] solamente me saca la columna nombre,  ["name"].nunique() le dice que saque los valores unicos en por nombre, es decir cuantas personas tienen un nombre unico - numero de elementos unicos; plot.bar() con esto le estoy diciendo  que me haga una grafica de barras-> en memoria genera todos los elementos para que solamente con un show() quede, 

```python
df.groupby("state")["name"].nunique().plot.bar()
plt.show()
```

el siguiente dataframe significa que se agrupe por estado y genero -> groupby(["state", "gender"]; .size() lo que hace es contar las catidades que aparecen segun la agrupacion anterior y los muestra 

```python
df.groupby(["state", "gender"]).size()
```

para graficarlo agrego .plot.bar(), para la grafica en cuention en valor del eje x = state y del eje Y= gender

```python
df.groupby(["state", "gender"]).size().plot.bar()
plt.show()
```

 se agrego la variable unstack() el cual aplila o unifica los valores de genero presentandolos en uno solo, con stacked=False se muestra en dos barras juntas, si en vez de false fuera True, se mostraria una barra sobre la otra de diferente color.

```python
df.groupby(["state", "gender"]).size().unstack().plot.bar(stacked=False)
plt.show()
```

```python
df.groupby(["gender","state"]).size().unstack().plot.bar(stacked=True)
plt.show()
```
