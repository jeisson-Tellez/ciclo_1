```python
import matplotlib.pyplot as plt
```

luego se  ponen las variables y los contenedores

```python
x1=[2,5,6,14]
y1= [11,22,33,44]

x2=[2,5,6,15]
y2=[4,12,18,45]

x3=list(range(16))
y3=list(map(lambda x: x**2,x3))
```

Lo que sigue, la aplicacion de la libreria, desde la terminal es imposible
dado que no imprime la grafica, debe estar en este tipo de archivos 

```python
plt.plot(x1,y1, color="blue", linewidth=3, label="linea 1")
plt.plot(x2,y2, color="green", linewidth=3, label="linea 2")
plt.plot(x3,y3,"o-", color="yellow", linewidth=1, label="linea 3")
plt.title("Grafica Ejemplo")
plt.xlabel("dinero gastado")
plt.ylabel("tiempo empleado")
plt.legend()
plt.grid()
plt.show()
```

Ahora bien, la estructura anterior de codigo tambien aplica para otro tipo de graficas, entonces, podemos encontrar
el grafico de barras usando el plt.bar

```python
plt.bar(x1,y1, color="red", linewidth=2, label=" primera linea")
plt.bar(x2,y2, color="green", linewidth=3, label=" segunda linea")
plt.title("Grafica Ejemplo")
plt.xlabel("dinero gastado")
plt.ylabel("tiempo empleado")
plt.legend()
plt.grid()
plt.show()
```

Ahora cambiamos de datos, para trabajar Histogramas, este lo usamos con plt.hist
esta grafica tiene otros argumentos, los principales son:
 datos: pueden estar ordenados o no, sin categorizacion o no, se entregan en una lista 
 Rangobin: son los rangos de analisis que manejan en los intervalos para el histograma; esto se le da en una variable que contenga una lista
 histtype: tipo de histograma que vamos hacer, "bar", es para el histograma de barras, se pueden usar otros 
 rwidth: este es el ancho 
 color: color de cada barra del histograma
 pero estos no son los unicos.
 


```python
Datos = [20,22,21,20,23,25,28,40,22,23,22,15,16,18,18,19,21,22,24,4,12,17,17,22,30,]
Rangobin=[0,10,20,20,30,40]

plt.hist(Datos,Rangobin,histtype="bar", rwidth=0.8, color="gray")
plt.title("Grafica Ejemplo")
plt.xlabel("dinero gastado")
plt.ylabel("tiempo empleado")
plt.yticks(range(0, 16, 1))
plt.show()

```

Entre tanto, tenemos las graficas de dispersion (x,y)
asi pues,  Scatter se pusa de la siguiente manera plt.scatter, sus argumentos son iguales a las lineas o a las barras 

```python
plt.scatter(x1,y1, color="blue", linewidth=3, label="linea 1")
plt.scatter(x2,y2, color="green", linewidth=3, label="linea 2")

plt.title("Grafica Ejemplo")
plt.xlabel("dinero gastado")
plt.ylabel("tiempo empleado")
plt.legend()
plt.grid()
plt.show()
```

entonces, seguimos con el grafico de pastel, que para el caso se llama pie y se usa con plt.pie, aqui se le pasa como arugmentos:
datos: en una lista
label: etiquetas, en una lista los str que  correspondan
colors: lista de colores que va a tomar la grafica de pastel 
startangle: algulo de inicio para iniciar a graficar la grafica pastel-  90 recomendado
shadow: esto habla de la sombra que tiene la grafica 
explode: resaltar uno de las porciones del grafico, para el caso de ejemplo explode=(0.1,0,0,0) resaltaria la primera porcion 
autopct: este me sirve para colocar los datos en procentajes de participacion de los datos %1.1f%%, con lo anterior podemos definir cuantos valores decimales queremos entre otras cosas 

```python
valores =[20,40,60,80]
plt.pie(valores, labels=["enero","febrero", "marzo", "abril"], colors=["red","orange","blue", "gray"],startangle=65,shadow=True,explode=(0.2,0,0.2,0), autopct="%1.1f%%" )
```

Ahora bien, para construir desde cero una grafica se usa la siguiente estructura:
1. plt.figure-> esto lo que hace es crear una hoja para generar nuestro grafico, el pixeles, pulgadas o proporcion
                 lo hace con el argumento figzise= (valor1, valor2), valor 2 es una proporcion de valor1, los dos iguales representan una cuadrado 
2. plt.subplot2grid-> representa la grilla  o rejilla que voy a usar, el numero de filas y colunmnas que va a tener la hoja y de esa manera como se distribuyen los espacios, seguido de la posicion del eje de la grafica en la hoja, inicia en (0,0)-> esto es una posicion que comienza en cero, entonces la forma (nfilas, ncolumnas) me dice cuantas graficas me van a salir en la  hoja que defini, comenzando la posicion 0,0 en la posicion superior izquierda.
                        subplot2grid((nfilas, ncolumnas)-> cantidad, (fila, columna)-> posicion)
3. colocamos los valores del dataframe que vamos a usar, pra el ejemplo -> df["survived"].value_counts() es la seleccion de informacion del dataframe, un subconjunto de este; lo siguiente -> plot.bar(alpha= 0.5), representa  que queremos dibujar <plot> un grafico de barra <.bar> con un alpha de 0.5, la cual representa la nitides de las barras
4. realizamos la siguiente grafica de informacion obtenida del dataframe, entonces, la siguiente expresion plt.subplot2grid((2,2),(0,1)) me dice que  me cree otra grafica de una hoja de 2 filas y 2 columnas y que me la ubique en la posicion de la grafica (0,0) fila cero, columna 1, seguido saco de nuevo la informacion df.survived.value_counts(normalize=True), se puede notar que el parametro normalize se pone como verdadero, lo cual indica que se quieren obtener porcentaje de los valores, seguido de la expresion .plot.bar(alpha=0.3)

5. las graficas de la columna sex, tienen igual tratamiento que las de survive, pero se ubican en posiciones diferentes

```python
plt.figure(figsize=(10,10))
plt.subplot2grid((2,2),(0,0))
df["survived"].value_counts().plot.bar(alpha= 1)
plt.title(" cantidad de sobrevivientes")
```

```python
plt.figure(figsize=(20,10))
plt.subplot2grid((2,2),(0,0))
df["survived"].value_counts().plot.bar(alpha= 1)
plt.title(" cantidad de sobrevivientes")


plt.subplot2grid((2,2),(0,1))
df["survived"].value_counts(normalize=True).plot.bar(alpha= 0.5)
plt.title(" procentaje de sobrevivientes")

plt.subplot2grid((2,2),(1,0))
df["sex"].value_counts().plot.bar(alpha= 1)
plt.title("cantidad segun sexo")

plt.subplot2grid((2,2),(1,1))
df["sex"].value_counts(normalize=True).plot.bar(alpha= 0.5)
plt.title("porcentaje segun sexo")
```
